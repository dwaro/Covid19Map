```python
import pandas as pandas
```

```python
df = pandas.read_csv("time_series_2019-ncov-Confirmed.csv")
```

```python
df
```

```python
formatted = []
```

```python
numCols = len(df.columns)
numCols
```

```python
def countVals(lst):
    lst = lst[4:]
    return lst[-1]
```

```python
globalCount = 0
countries = {}
for index, row in df.iterrows():
    total = countVals(df.loc[index])
    globalCount += total
    if (row["Province/State"] != row["Province/State"]):
        location = {"name": row["Country/Region"], "lat":row["Lat"], "lon": row["Long"], "country": row["Country/Region"], "total": total, "data": row[numCols-5:]}
        formatted.append(location)
    else:
        if row["Country/Region"] in countries:
            countries[row["Country/Region"]] += total
        else:
            countries[row["Country/Region"]] = total
        location = {"name": row["Province/State"], "lat":row["Lat"], "lon": row["Long"], "country": row["Country/Region"], "total": total, "data": row[numCols-5:]}
        formatted.append(location)
```

```python
countries
```

```python
newdf = pandas.DataFrame(formatted)
```

```python
newdf = newdf.sort_values(by ='total', ascending=False )
```

```python
cols = ['total']
newdf[cols] = newdf[newdf[cols] > 0][cols]
newdf = newdf.dropna(subset=["total"])
pandas.options.display.float_format = '{:,.0f}'.format
newdf
```

```python
globalCount
```

```python
newdf.to_json(r'total_cases.js', orient="records")
```
